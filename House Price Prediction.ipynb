```python
test_data = X_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

# Combine training and test data for consistent encoding
combined_data = pd.concat([X_train, X_test])
combined_data = pd.get_dummies(combined_data, columns=['ocean_proximity'])

# Split the combined data back into training and test sets
X_train_encoded = combined_data.iloc[:len(X_train), :]
X_test_encoded = combined_data.iloc[len(X_train):, :]

# Train the model
reg = LinearRegression()
reg.fit(X_train_encoded, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Log transform certain columns in both training and test sets
for col in ['total_rooms', 'total_bedrooms', 'population', 'households']:
    X_train[col] = np.log(X_train[col] + 1)
    X_test[col] = np.log(X_test[col] + 1)

# Combine training and test data for consistent encoding
combined_data = pd.concat([X_train, X_test])
combined_data = pd.get_dummies(combined_data, columns=['ocean_proximity'])

# Split the combined data back into training and test sets
X_train_encoded = combined_data.iloc[:len(X_train), :]
X_test_encoded = combined_data.iloc[len(X_train):, :]

# Fit the RandomForestRegressor model
forest = RandomForestRegressor()
forest.fit(X_train_encoded, y_train)

# Evaluate the model
score = forest.score(X_test_encoded, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {
    "n_estimators": [3, 10, 30],
    "max_features": [8, 12, 20],
    "min_samples_split": [2, 4, 6, 8]
}

# Assuming forest is a defined RandomForestRegressor instance
forest = RandomForestRegressor()

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring="neg_mean_squared_error", return_train_score=True)
grid_search.fit(X_train_encoded, y_train)

# Get best parameters and score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print(f"Best Parameters: {best_params}")
print(f"Best Score: {best_score}")
```
